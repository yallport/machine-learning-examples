```python
# importing libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib notebook

# used kaggle's Python Data Visualizations as tutorial - https://www.kaggle.com/benhamner/python-data-visualizations/notebook
```

```python
# loading the iris dataset, saved in the same directory
# the dataset is now a Pandas DataFrame

iris = pd.read_csv("Iris.csv")
```

```python
iris.head
```

```python
# how many examples we have for each species

iris["Species"].value_counts()
```

```python
# using .plot from Pandas to use a scatterplot of the iris features

iris.plot(kind = "scatter", x = "SepalLengthCm", y = "SepalWidthCm")
```

```python
# using seaborn to make a similar plot
# shows both scatterplot and histogram in the same figure

sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", data = iris, size = 5)
```

```python
# above plot shows length and width, but not species
# use FacetGrid to color the scatterplot by species

sns.FacetGrid(iris, hue = "Species", size = 5) \
    .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
    .add_legend()
```
